## Cleaning

 ### Quality issues :
#### In the dataframe from twitter archive (df1):
 - some have missing urls.

 - missing names such as a , an , none , the

#### In dataframe from the image_preditictions.tsv file (df2):
 - The tweet_id erronous datatype . 
 



#### In dataframe from twitter API (df3):
 - id is tweet_id in all otherdata sets except this one and has errornous data type.
 - undescriptive language tags .
 
 
 
 
### Tideness issues :
 - doggo, floofer , pupper have missing values could be melted into one in df1 .

```python
df1_c= df1.copy()
df2_c=df2.copy()
df3_c = df3.copy()
```

```python
# Remove the values from the dataframe which doesn't have a url .

# code:

df1_c= df1_c[df1_c.expanded_urls.notna()]
```

```python
# test : 

df1_c.info()
```

```python
# change the datatypes of errornous data in df1 : 

# code :
df1_c.tweet_id= df1_c.tweet_id.astype('object')
df1_c.in_reply_to_status_id= df1_c.in_reply_to_status_id.astype('object')
df1_c.in_reply_to_user_id = df1_c.in_reply_to_user_id .astype('object')
df1_c.retweeted_status_user_id = df1_c.retweeted_status_user_id .astype('object')
df1_c.retweeted_status_id = df1_c.retweeted_status_id.astype('object')
```

```python
#test : 
df1_c.info()
```

```python
# change the invalid name dogs to none . 

#code :
df1_c['name'].replace('the', 'None', inplace=True)
df1_c['name'].replace('an' , 'None', inplace=True)
df1_c['name'].replace('a', 'None', inplace=True)
df1_c['name'].replace('quite', 'None', inplace=True)
```

```python
# test :
df1_c.name.value_counts()
```

```python
# change all the invalid denominator ratings to 10 

# code :
invalid = df1_c.rating_denominator[df1_c.rating_denominator != 10]
for i in invalid :
    df1_c['rating_denominator'].replace( i , 10 , inplace=True)
    
```

```python
#test :
df1_c[df1_c.rating_denominator != 10]
```

```python
#remove retweeted stuff from dataframe :

#code : 
df1_c = df1_c[~df1_c.retweeted_status_id.notna()]
df1_c .drop(['retweeted_status_id' ,'retweeted_status_user_id' , 'retweeted_status_timestamp'] , axis=1,inplace=True)


#test :
df1_c.info()
```

```python
# change the datatype of tweet_id in df2 to object 


#code : 
df2_c.tweet_id= df2_c.tweet_id.astype('object')
```

```python
#test:
df2_c.info()
```

```python
#change df3 id to tweet_id and change its data type :

#code : 
df3_c = df3_c.rename(columns={'id': 'tweet_id'})
df3_c.tweet_id = df3_c.tweet_id.astype('object')
```

```python
#test:
df3_c.info()
```

```python
# change all non english langugages in en to not english

# code :
wrong = df3_c.lang[df3_c.lang != 'en']
for w in wrong : 
    df3_c['lang'].replace(wrong , 'not english ', inplace=True)
```

```python
#test :
df3_c.lang.value_counts()
df1_c.info()
```

```python
#tideness issue :
#melt in df1 dataframe for doggo,pupper,floffer,pupper,puppo

#code:
stage = []
df1_c['growth_stage']=df1_c.text.str.extract('(doggo|pupper|floofer|puppo)' , expand = True)
  
```

```python
#test : 
df1_c.sample(10)
```

```python
# drop the puppo , pupper , floofer , doggo column now 

#code : 
df1_c = df1_c.drop(['doggo', 'floofer', 'pupper', 'puppo' ] , axis = 1  )
```

```python
#test : 
df1_c.info()
```

```python
# combine the three tables 

#code 
master = pd.merge (df1_c , df2_c , on =['tweet_id'] , how = 'inner')
```

```python
master.info()
```

```python
master.sample(5)
```

```python
final = pd.merge(master, df3_c , on= ['tweet_id'] , how = 'inner')
```

```python
#test 
final.info()
```
